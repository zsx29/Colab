```python
# csv 파일 데이터프레임 불러오기
df_exam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/exam.csv')
df_exam
```

```python
# 기본 요약 통계
df_exam.describe()
```

```python
# 데이터프레임 인덱싱
# loc - 컬럼명 인덱싱
df_math = df_exam.loc[:, "math"]
print(df_math)

df_english = df_exam.loc[:, 'english']
df_english

df_rs = df_exam.loc[10:, ["math", "english", "science"]]
df_rs
# iloc - 인덱스 인덱싱

df_rs2 = df_exam.iloc[1:5, 2:4]  # 1 ~ 4 행, 2 ~ 3열 까지
df_rs2
```

```python
# 수학점수 60점 이상 데이터 프레임
df_exam[df_exam.math >= 60]
```

```python
# 수학점수 내림차순
df_exam.sort_values(by="math", ascending=False)
```

```python
# 영어점수 60 ~ 70, 내림차순
df_exam[(df_exam.english >= 60) & (df_exam.english < 70)].sort_values(by="english", ascending="F")
```

```python
# total
df_exam["total"] = df_exam.math + df_exam.english + df_exam.science
df_exam

df_exam["mean"] = df_exam.total / 3
df_exam
```
